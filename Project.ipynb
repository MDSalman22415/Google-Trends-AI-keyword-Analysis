```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import plotly.express as px
from pytrends.request import TrendReq
```

# Setup pytrend library and keyword define

```python
pyternds = TrendReq(hl='eg-Us')
keyword = 'ai'
```

# Data Request

```python
pyternds.build_payload([keyword],cat=0,timeframe='today 12-m',geo='',gprop='')
```

# country wise interest

```python
region_data = pyternds.interest_by_region()
region_data = region_data.sort_values(by=keyword,ascending=False).head(15)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=region_data[keyword],y=region_data.index,palette='Blues_d')
plt.title(f"To 10 country serching this '{keyword}' ")
plt.xlabel('Interset')
plt.ylabel('Country')
plt.show()
```

# World map

```python
region_data = region_data.reset_index()
fig = px.choropleth(region_data,
                   locations='geoName',
                   locationmode='country names',
                   color = keyword,
                   title=f"Serch interest for '{keyword}' ",
                   color_continuous_scale='Blues')
fig.show()
```

# time wise keyword

```python
time_df = pyternds.interest_over_time()
```

```python
plt.figure(figsize=(12,6))
plt.plot(time_df.index, time_df[keyword],marker ='o',color = 'blue')
plt.title(f"Search interest over time '{keyword}' ")
plt.xlabel=('Date')
plt.ylabel("Interest")
plt.grid(True)
plt.show()
```

# multiple keyword compare

```python
kw_list = ['cloud computing','data science','machine learning']
pyternds.build_payload(kw_list,cat=0,timeframe='today 12-m',geo='',gprop='')
```

```python
compare_df = pyternds.interest_over_time()
plt.figure(figsize=(12,6))
for kw in kw_list:
    plt.plot(compare_df.index, compare_df[kw],label = kw)

plt.title = ("Keyword comparison over time")
plt.xlabel = ("Date")
plt.ylabel("Interest")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
